Do Feature Selection

```python
#create a new dataframe only for the necessary features

movies = df[['id', 'title', 'overview', 'genre']]
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genre']
movies
```

```python
movies.drop(columns=['overview', 'genre'], inplace=True)
movies
```

```python
movies.isnull().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=1000, stop_words='english')
cv
```

```python
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity
```

```python
movies[movies['title']=='The Godfather'].index[0]
```

```python
movies.info()
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
  print(movies.iloc[i[0]].title)
```

```python
def recommend(movie):
    index = movies[movies['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
      print(movies.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
